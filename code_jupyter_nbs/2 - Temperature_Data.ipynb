## Converting into long form

```python
# Generating a long format for temperature 
temperature_long = temperature.melt(
                id_vars=["station_id","year","month","day"],
                value_vars=[f"h{i}" for i in range(1, 25)],
                var_name="hour",
                value_name="temperature"
                        )

temperature_long.head()
```

```python
# Generating timestamp date from the available time information

temperature_long["date"]=pd.to_datetime(
    temperature_long["year"].astype(str) + "-" +
    temperature_long["month"].astype(str).str.zfill(2) + "-" +
    temperature_long["day"].astype(str).str.zfill(2) + " " 
)

temperature_long.head()
```

```python
# Mapping hour values
hour_map = {f"h{i}": f"{str(i-1).zfill(2)}:30" for i in range(1, 25)}

# Replace the values using .map()
temperature_long["hour"] = temperature_long["hour"].map(hour_map)
temperature_long.head()
```

```python
# Creating timestamp from available time information

temperature_long["timestamp"] = pd.to_datetime(
    temperature_long["year"].astype(str) + "-" +
    temperature_long["month"].astype(str).str.zfill(2) + "-" +
    temperature_long["day"].astype(str).str.zfill(2) + " " +
    temperature_long["hour"]
)

temperature_long = temperature_long.sort_values(by="timestamp").reset_index(drop=True)
temperature_long.head()
```

```python
# Removing columns and re-ordering

temperature_long = temperature_long[["station_id", "temperature", "timestamp"]]
temperature_long = temperature_long.sort_values(by=["timestamp", "station_id"]).reset_index(drop=True)
temperature_long.head()
```

## Converting into a matrix
Developing a timestamp (row) vs station (column) matrix

```python
# Using pivot on timestamp as index
temperature_matrix = temperature_long.pivot(index="timestamp", columns="station_id", values="temperature")
temperature_matrix.head()
```

Now for every timestamp, a linear combination of temperature from different station ids can be used.

## PCA
PC1 can capture general temperature reading across stations, while subsequent PCs can capture factors 'other' than main temperature signal.
<br>
These 'factors' could be elevation, proximity to ocean, etc. that might be explaining the secondary variation in temperature readings.

```python
# Calculating variance for each station_id
import numpy

for i in range(11):
    print(f'Variance for station {i+1} is: {numpy.nanvar(temperature_matrix.iloc[:,i])}')
```

Variance Range: 279-322 <br>
Variation of temperatures within each station appear more or less similar. Hence, normalisation is not required.

```python
# Temperature df without forecasting horizon
temperature_no_na = temperature_matrix.loc[:'2008-6-30 05:30:00',:]
```

```python
# performing PCA
from sklearn.decomposition import PCA

temp_pca = PCA().fit(temperature_no_na)
```

[Reference for PCA](https://www.geo.fu-berlin.de/en/v/soga-py/Advanced-statistics/Multivariate-Approaches/Principal-Component-Analysis/PCA-in-Python/index.html)

```python
# Eigenvector Matrix
print(f"Rotation matrix:\n{temp_pca.components_}")
```

```python
import numpy as np
print(f"Standard deviation of each principal component:\n{np.sqrt(temp_pca.explained_variance_)}")
```

```python
print(f"Proportion of variance explained by each principal component:\n{np.round(temp_pca.explained_variance_ratio_, 3)}")
```

PC1 is most essential and explains 97% of total variation in temperature. PC2 only explains about 1%.

Hence a single PC can strongly capture the broad temperature signal, with assigning suitable weights that maximise explained variance.

```python
# Weight assignment to each station by PC 1
print(temp_pca.components_[0])
print(f'Sum of weights: {sum(temp_pca.components_[0])}')
```

Weights are actually closer together in sign and magnitude. This suggests that there are no stations that are contrasting with others or have extreme readings, most of them appear to move together.

```python
# Re-scaling weights
weights_normalised = temp_pca.components_[0]/sum(temp_pca.components_[0])
print(weights_normalised)
print(f'Sum of normalised weights: {sum(weights_normalised)}')
```

```python
# Matrix Multiplication using normalised weights
temperature_no_na["temp_weighted"] = temperature_no_na.dot(weights_normalised)
temperature_no_na.head()
```

```python
# Exporting
temperature_no_na.to_csv("weighted_temperature.csv", index=True)
```
